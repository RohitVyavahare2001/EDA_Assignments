# Question 1



Write a program to load penguins dataset from seaborn and plot histplot/displot for the flipper_length_mm.

# Question 14


Write a program to load tips dataset from seaborn and plot regression line between total_bill and tip with 95% confidence interval.

```python
import seaborn as sns
tips=sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips, ci=95)
```

# Question 15
